```python
import cv2 as cv2
from matplotlib import pyplot as plt
import pandas
import json
```

```python
id = 1269
```

```python
train_data = pandas.read_csv("./debris-detection/train.csv")

test_data = pandas.read_csv("./debris-detection/test.csv")
```

```python
train_bboxes = json.loads(train_data["bboxes"][id])
test_bboxes = json.loads(test_data["bboxes"][id])

train = True

image = cv2.imread(
    f"./debris-detection/train/{id}.jpg"
    if train
    else f"./debris-detection/test/{id}.jpg"
)
```

```python
for bbox in train_bboxes if train else test_bboxes:
    if train:
        x1, x2, y1, y2 = bbox
    else:
        x1, y1, x2, y2, prob = bbox
        x1, x2, y1, y2 = int(x1), int(x2), int(y1), int(y2)
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
```

```python
edges = cv2.Canny(image, 200, 250)
```

```python
plt.imshow(edges)
```
